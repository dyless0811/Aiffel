multiprocessing.Pool.mapì„ í†µí•´ ì—¬ëŸ¬ ê°œì˜ í”„ë¡œì„¸ìŠ¤ì— íŠ¹ì • í•¨ìˆ˜ë¥¼ ë§¤í•‘í•´ì„œ ë³‘ë ¬ì²˜ë¦¬í•˜ë„ë¡ êµ¬í˜„í•˜ëŠ” ë°©ë²•ì´ ë„ë¦¬ ì‚¬ìš©ë©ë‹ˆë‹¤.

```python
from multiprocessing import Pool
from os import getpid

def double(i):
    print("I'm processing ", getpid())    # pool ì•ˆì—ì„œ ì´ ë©”ì†Œë“œê°€ ì‹¤í–‰ë  ë•Œ pidë¥¼ í™•ì¸í•´ ë´…ì‹œë‹¤.
    return i * 2

with Pool() as pool:
      result = pool.map(double, [1, 2, 3, 4, 5])
      print(result)
```

ìœ„ì—ì„œ double(i)ì´ë¼ëŠ” ë©”ì†Œë“œê°€ poolì„ í†µí•´ ê°ê° ë‹¤ë¥¸ pidë¥¼ ê°€ì§„ í”„ë¡œì„¸ìŠ¤ë“¤ ìœ„ì—ì„œ multiprocessë¡œ ì‹¤í–‰ë˜ì—ˆë‹¤ëŠ” ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
import math
import concurrent
import time

# ì†Œìˆ˜(prime) íŒë³„ ë¬¸ì œ
PRIMES = [
    112272535095293,
    112582705942171,
    112272535095293,
    115280095190773,
    115797848077099,
    1099726899285419]

# ì†Œìˆ˜ íŒë³„ í•¨ìˆ˜
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    sqrt_n = int(math.floor(math.sqrt(n)))
    for i in range(3, sqrt_n + 1, 2):
        if n % i == 0:
            return False
    return True

# ì†Œìˆ˜ íŒë³„ í•¨ìˆ˜ í˜¸ì¶œ
def main():
    # ë³‘ë ¬ì²˜ë¦¬ ìˆ˜í–‰
    print("ë³‘ë ¬ì²˜ë¦¬ ì‹œì‘")
    start = time.time()
    # concurrent.futures ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ í”„ë¡œì„¸ìŠ¤ í’€ì—ì„œ ë™ì‘í•˜ê²Œ í•˜ê¸° ìœ„í•´ with ë¬¸
    with concurrent.futures.ProcessPoolExecutor() as executor:
        # ë§µ-ë¦¬ë“€ìŠ¤(map-reduce)ìŠ¤íƒ€ì¼, map() í•¨ìˆ˜ë¥¼ ProcessPoolExecutor() ì¸ìŠ¤í„´ìŠ¤ì—ì„œ ìƒì„±ëœ executor ì—ì„œ ì‹¤í–‰
        for number, prime in zip(PRIMES, executor.map(is_prime, PRIMES)):
            print('%d is prime: %s' % (number, prime))
    end = time.time()
    print("ë³‘ë ¬ì²˜ë¦¬ ìˆ˜í–‰ ì‹œê°", end-start, 's', '\n')

    # ë‹¨ì¼ì²˜ë¦¬ ìˆ˜í–‰
    print("ë‹¨ì¼ì²˜ë¦¬ ì‹œì‘")
    start = time.time()
    for number, prime in zip(PRIMES, map(is_prime, PRIMES)):
        print('%d is prime: %s' % (number, prime))
    end = time.time()
    print("ë‹¨ì¼ì²˜ë¦¬ ìˆ˜í–‰ ì‹œê°", end-start, 's')
            
if __name__ == '__main__':
    main()
```
